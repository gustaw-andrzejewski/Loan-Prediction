```python
#now we will prepare the data to be trained 
```

```python
#changing categorical variables to numerical
d = {'Yes':1, 'No': 0}
df_train['Married'] = df_train['Married'].map(d)
df_train['Self_Employed'] = df_train['Self_Employed'].map(d)
d = {'Graduate': 1, 'Not Graduate':0}
df_train['Education'] = df_train['Education'].map(d)
d = {'Male':1, 'Female':0}
df_train['Gender'] = df_train['Gender'].map(d)
d = {'Y':1, 'N':0}
df_train['Loan_Status'] = df_train['Loan_Status'].map(d)
d = {'Urban':2,'Semiurban':1, 'Rural': 0}
df_train['Property_Area'] = df_train['Property_Area'].map(d)
```

```python
#checking value types
df_train.info()
```

```python
df_train['Dependents'] = pd.to_numeric(df_train['Dependents'], errors = 'coerce')
df_train.info()
```

```python
#checking missing values
df_train.isnull().sum()
```

```python
#for categorical variables we will use mode to fill the lacking values
df_train['Gender'].fillna(df_train['Gender'].mode()[0], inplace=True)
df_train['Married'].fillna(df_train['Married'].mode()[0], inplace=True)
df_train['Dependents'].fillna(df_train['Dependents'].mode()[0], inplace=True)
df_train['Self_Employed'].fillna(df_train['Self_Employed'].mode()[0], inplace=True)
df_train['Credit_History'].fillna(df_train['Credit_History'].mode()[0], inplace=True)
```

```python
df_train.isnull().sum()
```

```python
#checking Loan_Amount_Term 
df_train['Loan_Amount_Term'].value_counts()
```

```python
#we will use mode to fill missing values because it is very frequent compared to the other 
df_train['Loan_Amount_Term'].fillna(df_train['Loan_Amount_Term'].mode()[0],inplace=True)
```

```python
#for the LoanAmount we will use median because of the distribution we saw on the plot earlier(the fact that there are outliers)   
df_train['LoanAmount'].fillna(df_train['LoanAmount'].median(),inplace=True)
df_train.isnull().sum()
```

```python
#dealing with outliers in LoanAmount by log transformation
df_train['LoanAmount'] = np.log(df_train['LoanAmount'])
df_train['LoanAmount'].hist()
```

```python
#preparing for train test split
x = df_train.drop(columns= 'Loan_Status')
y = df_train['Loan_Status']
```

```python
x.shape
```

```python
y.shape
```

```python
#splitting data into train and test 
import sklearn
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
#making a logistic regression model
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=1)
lr.fit(x_train, y_train)
```

```python
#measuring the accuracy of the model
from sklearn.metrics import accuracy_score

pred_test = lr.predict(x_test)
accuracy_score(y_test, pred_test)
```
